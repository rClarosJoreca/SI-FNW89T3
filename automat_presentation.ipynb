```python
import os
import glob
import pandas as pd
from pptx import Presentation
from pptx.dml.color import RGBColor
```

```python
prs = Presentation("2409_Auto_France_September 24_Leboncoin.pptx")
```

```python
for idx, diapositiva in enumerate(prs.slides):
    print(f"\n--- Diapositiva {idx + 1} ---")
    
    # Iterar sobre cada objeto (shape) en la diapositiva
    for shape in diapositiva.shapes:
        
        # Mostrar el tipo de shape
        print(f"Objeto tipo: {shape.shape_type}")
        
        # Si el objeto tiene un cuadro de texto
        if shape.has_text_frame:
            print("Contiene texto:")
            print(f"Texto actual: {shape.text}")
        
        # Si el objeto es una imagen (tipo Picture)
        if shape.shape_type == 13:  # 13 es el tipo para imágenes
            print(f"Este es una imagen.")
        
        # Si el objeto es un gráfico
        if shape.has_chart:
            print(f"Este es un gráfico.")
            
        # Si el objeto es un marcador de posición
        if shape.is_placeholder:
            print(f"Este es un marcador de posición: {shape.placeholder_format.idx}")
```

```python
from pptx.chart.data import CategoryChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.enum.chart import XL_LEGEND_POSITION
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor 


slide = prs.slides[3]
slide_layout = slide.slide_layout
new_slide = prs.slides.add_slide(slide_layout)


title = new_slide.shapes.title
title.text = "Auto FR classifieds market overview | September 2024"


chart_data = CategoryChartData()
chart_data.categories = ['Pro Listings', 'Private Listings']  
chart_data.add_series('Listings', (50, 50)) 


x, y, cx, cy = Inches(2), Inches(1.5), Inches(6), Inches(4.5)
chart = new_slide.shapes.add_chart(
    XL_CHART_TYPE.DOUGHNUT, x, y, cx, cy, chart_data
).chart


chart.has_legend = True 
chart.legend.position = XL_LEGEND_POSITION.BOTTOM 
chart.plots[0].has_data_labels = True  
chart.plots[0].data_labels.number_format = '0%'  


series = chart.series[0]

# Cambiar el color del primer punto (Pro Listings)
fill = series.points[0].format.fill
fill.solid()
fill.fore_color.rgb = RGBColor(0x4B, 0x00, 0x4B)  # Morado oscuro para "Pro Listings"

# Cambiar el color del segundo punto (Private Listings)
fill = series.points[1].format.fill
fill.solid()
fill.fore_color.rgb = RGBColor(0xFF, 0xD7, 0x00)  # Amarillo para "Private Listings"

# Modificar las propiedades de las etiquetas de datos
data_labels = chart.plots[0].data_labels
data_labels.font.size = Pt(12)  # Tamaño de la fuente para las etiquetas
data_labels.font.color.rgb = RGBColor(0x00, 0x00, 0x00)  # Color negro para el texto

# Añadir cuadro de texto en el centro del gráfico para simular la etiqueta grande "1.14 M"
txBox = new_slide.shapes.add_textbox(Inches(3.5), Inches(2.75), Inches(2), Inches(1))
tf = txBox.text_frame
p = tf.add_paragraph()
p.text = "1.14 M"
p.font.size = Pt(28)
p.font.bold = True
p.font.color.rgb = RGBColor(0x00, 0x00, 0x00)  # Texto negro

# Añadir un cuadro de texto para el porcentaje de cambio "-5.9% vs August-24"
txBox_change = new_slide.shapes.add_textbox(Inches(3.5), Inches(3.25), Inches(2), Inches(1))
tf_change = txBox_change.text_frame
p_change = tf_change.add_paragraph()
p_change.text = "-5.9% vs August-24"
p_change.font.size = Pt(14)
p_change.font.bold = True
p_change.font.color.rgb = RGBColor(0xFF, 0x00, 0x00)  # Texto en rojo

# Guardar la presentación con el gráfico personalizado
prs.save("presentacion_modificada_con_grafico_estilo_personalizado.pptx")
print("Presentación guardada con un gráfico de donut personalizado.")
```

```python
# Step 1: Add a new slide at the end (this is required as python-pptx always appends)
slide_layout = prs.slide_layouts[1]  # Example layout (title and content)
new_slide = prs.slides.add_slide(slide_layout)

# Step 2: Add content to the new slide
title = new_slide.shapes.title
title.text = "Inserted Slide"

# Step 3: Reorder slides to insert the new slide between two specific slides
xml_slides = prs.slides._sldIdLst  # Access the list of slides in XML
slides = list(xml_slides)
slide_to_insert = slides[-1]  # Get the slide that was just added (last slide)

# Insert the slide at position 2 (between the 1st and 2nd slides, for example)
xml_slides.remove(slide_to_insert)
xml_slides.insert(1, slide_to_insert)  # Insert after the first slide

# Save the modified presentation
prs.save("presentacion_con_diapositiva_insertada.pptx")
print("Presentación guardada con una diapositiva insertada.")
```
